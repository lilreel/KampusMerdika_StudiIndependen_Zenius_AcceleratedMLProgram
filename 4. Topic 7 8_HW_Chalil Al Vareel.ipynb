#### Nomor 2a:

```python
#Buat fungsi menghitung rata-rata dari Annual Price Berdasarkan jumlah kamar per rumahnya (No_Rooms)
def kolum_y(no_room, ann_price):
    for i in no_room:
        sum_of_list = 0
        for j in range(len(ann_price)):
            sum_of_list += ann_price[j]
        average = sum_of_list/len(ann_price)
        return average
```

```python
#Membuat Variabel yang menghitung rata-rata dari Annual Price Berdasarkan jumlah kamar per rumahnya (No_Rooms)
df2['AnnualPrice/No_Room']=df2.apply(lambda x : kolum_y(df2['No_Rooms'], df2['AnnualPrice']), axis=1)

#Visualisasi menggunakan boxplot
pd.crosstab(df2['Bathroom'], df2['AnnualPrice/No_Room']).plot(kind = 'bar', stacked = True)
plt.show()
```

#### Nomor 3a:

```python
x = df3["body_mass_g"]
y1 = df3["flipper_length_mm"]
y2 = df3["bill_length_mm"]

colors = df3["island"]
```

```python
fig, (ax1,ax2) = plt.subplots(2, figsize=(10,6))

ax1.set_title('Flipper Length based on Body Mass')
sns.scatterplot(x=x, y=y1, data=df2, hue = colors, ax=ax1)
ax1.set_xlabel('Body Mass')
ax1.set_ylabel('Flipper Length')

ax2.set_title('Bill Length based on Body Mass')
sns.scatterplot(x=x, y=y2, data=df2, hue = colors, ax=ax2)
ax2.set_xlabel('Body Mass')
ax2.set_ylabel("Bill Length")

plt.tight_layout()
plt.show()
```
