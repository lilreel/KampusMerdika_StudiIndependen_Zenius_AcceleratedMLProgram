```python
import pandas as pd
import numpy as np
```

1. Create Pandas DataFrame

```python
#create DataFrame
bmi_df = pd.DataFrame({
    'name' : ['Anna', 'Dane', 'David', 'Kevin', 'Joe', 'Rose'],
    'weight' : [45.2, 50.5, 62.1, 64.0, 59.0, np.nan],
    'height' : [155, 160, 162, 170, 167, 159],
    'birth_year' : [2000, 1997, 1996, 1987, 1998, 2003],
    'gender' : ['female', 'male', 'male', 'male', 'male', 'female'],
    'city' : ['New York', 'Chicago', 'Austin', 'New York', np.nan, 'New York']
})
bmi_df
```

2. Create another Pandas DataFrame as the following and merge it with the previous 
DataFrame (key=name and nick name). Make sure that no rows are missing from the 
original DataFrame

```python
#Another dataFrame
bmi_df2 = pd.DataFrame({
    'nick_name' : ['Anna', 'David', 'Dane', 'Rose'],
    'full_name' : ['Anna Levis', 'David John', 'Dane Sebastian', 'Rose Kimberly']
})
bmi_df2
```

```python
#Merging as Outer-Join
bmi_df3 = bmi_df.merge(bmi_df2,  left_on = "name", right_on = "nick_name", how = 'outer')
bmi_df3
```

3. Drop nick_name column and fill null values in the full_name from the merged 
DataFrame with their original name column.


```python
#Drop nick_name column
bmi_df4 = bmi_df3.drop(columns = 'nick_name')
bmi_df4
```

```python
#Fill null values in the full_name with column name
bmi_df4['full_name'] = bmi_df4['full_name'].fillna(bmi_df4['name'])
bmi_df4
```

4. Create a new ‘bmi’ column calculated based on the following formula<br><br>
 **BMI (kg/m^2) = weight / (height/100)^2**

```python
#BMI Calculation
bmi_df4['bmi'] = bmi_df4['weight']/(np.square(bmi_df4['height']/100))
bmi_df4
```

5. Create a new ‘bmi_class’ column following the definition in the table below

```python
#kelas pendefinisian bmi
def classBMI(val):
  stats = ''
  if val<18.5 :
    stats = 'Underweight'
  elif val<25 :
    stats = 'Normal Weight'  
  elif val<30 :
    stats = 'Overweight'  
  elif val<35 :
    stats = 'Obesity Class 1'  
  elif val<40 :
    stats = 'Obesity Class 2'  
  elif val>=40 :
    stats = 'Obesity Class 3'  
  else :
    stats = 'Unknown'
  return stats

#dataframe
bmi_df4['bmi_class'] = bmi_df4['bmi'].apply(classBMI)
bmi_df4
```
